##PySpark functions to perform proyect operations

```python
from pyspark.sql.functions import col

# 1) Customer Service jobs in the United States
cs_us = cs_jobs.filter(col("search_country") == "United States")
cs_us.select("job_title", "company", "search_country").show(5, truncate=False)

# 2) Customer Service jobs in Canada
cs_ca = cs_jobs.filter(col("search_country") == "Canada")
cs_ca.select("job_title", "company", "search_country").show(5, truncate=False)

# 3) Full-time Customer Service jobs (single condition on job_type)
cs_full_time = cs_jobs.filter(col("job_type") == "FULL_TIME")
cs_full_time.select("job_title", "job_type").show(5, truncate=False)

# 4) Remote OR Hybrid Customer Service jobs (OR condition)
cs_remote_or_hybrid = cs_jobs.filter(
    (col("job_type") == "REMOTE") | (col("job_type") == "HYBRID")
)
cs_remote_or_hybrid.select("job_title", "job_type", "job_location").show(5, truncate=False)

# 5) Customer Service jobs that ALREADY have a summary and are being worked on (AND)
# (in your data got_summary / is_being_worked look like 't'/'f' strings)
cs_active = cs_jobs.filter(
    (col("got_summary") == "t") & (col("is_being_worked") == "t")
)
cs_active.select("job_title", "company", "got_summary", "is_being_worked").show(5, truncate=False)
```

##H. Aggregate window functions

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import sum as _sum

# Window ordenada por skill_count descendente
w_cum = Window.orderBy(col("skill_count").desc()) \
    .rowsBetween(Window.unboundedPreceding, 0)

skills_with_cum = skill_counts.withColumn(
    "cumulative_skill_count",
    _sum("skill_count").over(w_cum)
)

skills_with_cum.orderBy(col("skill_count").desc()).show(20, truncate=False)
```

What this window does

It orders skills from the most frequent to the least frequent.

It calculates a progressive total, adding each skill’s count to all previous ones.

The window "unboundedPreceding → current row" means:

“Start from the very first skill (rank 1) and sum everything up to the current skill.”

Why this matters  ???

This tells us:

How much demand is explained by the top N skills.

Whether the market is dominated by a few key capabilities.

Whether skill requirements are diversified or concentrated.

Query 2

```python
from pyspark.sql.functions import col

# Window que incluye todas las filas
w_all = Window.rowsBetween(Window.unboundedPreceding, Window.unboundedFollowing)

skills_with_share = skills_with_cum.withColumn(
    "total_skill_count",
    _sum("skill_count").over(w_all)
).withColumn(
    "skill_share",
    col("skill_count") / col("total_skill_count")
)

skills_with_share.orderBy(col("skill_count").desc()).show(20, truncate=False)
```

Why this matters

It normalizes skill demand.

It lets us say “Skill X represents Y% of all required skills.”

It provides a clean comparison across skills with different counts.

Example interpretation:

“Communication represents 12% of all skills required in Customer Service roles.”

This is important because percentages communicate the importance of each skill clearly and intuitively.
